### Time Based Analysis

```python
data['Year']=pd.to_datetime(data['Year'])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
data['Year'].value_counts().sort_index().plot(kind="line",color="blue",marker="o",linewidth=2)
plt.title("Earthquake count per year",fontweight="bold",fontsize=14)
plt.xlabel("Year",fontweight="bold")
plt.ylabel("Number of Earthquakes",fontweight="bold")
plt.grid(True)
plt.show()
```

### Magnitude Analysis

```python
import matplotlib .pyplot as plt
plt.figure(figsize=(6,4))
plt.hist(data["magnitude"],color="purple",edgecolor="black",bins=15)
plt.hist(data['magnitude'])
plt.title("Magnitude Distribution",fontweight="bold",fontsize=14)
plt.xlabel("Magnitude",fontweight="bold")
plt.ylabel("Frequency",fontweight="bold")
plt.grid(True)
plt.show()
```

#### Box Plot

```python
import seaborn as sns
import matplotlib .pyplot as plt 
plt.figure(figsize=(6,4))
sns.boxplot(x=data['magnitude'],color="orange")
plt.title("Magnitude Analysis(boxplot)",fontweight="bold",fontsize=14)
plt.xlabel("Magnitude",fontweight="bold")
plt.grid(True)
plt.show()
```

### Depth Analysis

```python
plt.figure(figsize=(6,4))
plt.hist(data['depth'],color="skyblue",edgecolor="black")
plt.title("Depth Distribution",fontweight="bold")
plt.xlabel("Depth",fontweight="bold")
plt.ylabel("Frequency",fontweight="bold")
plt.show()
```

### Tsunami VS Non-Tsunami Comparison

```python
import seaborn as sns
import matplotlib .pyplot as plt
plt.figure(figsize=(6,4))
sns.boxplot(x="tsunami",y="magnitude",data=data,palette="Set2")
plt.title("Tsunami VS Non-Tsunami Comparison",fontweight="bold")
plt.xlabel("longitude",fontweight="bold")
plt.ylabel("latitude",fontweight="bold")
plt.show()
```

### Location Plot (Latitude VS Longitude)

```python
plt.figure(figsize=(6,4))
plt.scatter(data['longitude'],data['latitude'],c=data["magnitude"],cmap="rainbow")
plt.title("Earthquake Locations")
plt.xlabel("longitude",fontweight="bold")
plt.ylabel("latitude",fontweight="bold")
plt.show()
```

#### With Tsunami Color

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="longitude",y="latitude",data=data,hue="magnitude",palette="rainbow")
plt.title("Earthquake Locations",fontsize=12,fontweight="bold")
plt.xlabel("longitude",fontweight="bold")
plt.ylabel("latitude",fontweight="bold")
plt.show()
```

### Correlation Heatmap

```python
corr=data.corr()
sns.heatmap(corr,annot=True)
plt.show()
```
